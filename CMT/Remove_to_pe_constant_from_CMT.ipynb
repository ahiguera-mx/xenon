```python
import straxen
import numpy as np
```

```python
st = straxen.contexts.xenonnt_online()
```

```python
my_gains = np.ones(494)*0.01
st.set_config(dict(gain_model=('CMT_model',("to_pe_constant",my_gains ))))
```

```python
peaks = st.get_array('009741','peaks')
```

* The option `st.set_config(dict(gain_model=('CMT_model',("to_pe_constant",my_gains ))))` throws the above error, the reason is straxen does not support arrays as option for the metadata, therefore this option should be removed from CMT <br>
* Instead the user should use the following option which is equivalent

```python
straxen.FIXED_TO_PE['my_gain_model']=my_gains
```

```python
st.set_config(dict(gain_model=('to_pe_constant','my_gain_model')))
```
